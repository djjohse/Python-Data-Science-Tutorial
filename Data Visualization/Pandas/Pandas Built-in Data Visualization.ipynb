## `.plot.line()`

A line plot requires that you specify which column you want to use for the y-axis.
The default for the x-axis is the index of the dataframe.

```python
df1.head()
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

## `.plot.scatter()`

Create a scatter plot with varying marker point size and color.

```python
df1.head()
```

We can plot **A vs B**, and graph the **C** value by how 'hot' the point is.

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

We can also plot the value of **C** by point size as shown below.

```python
df1.plot.scatter(x='A',y='B',s=100*df1['C'])
```

## `.plot.hexbin()`

```python
dfn = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
dfn.head()
```

`.hexbin()` can plot a density heat map. In this case, you can see that the hex spaces with more values are colored darker. Since these numbers were randomly generated from a normal distribution, the count per hex (color) is grouped darker around the center.

```python
dfn.plot.hexbin(x='a',y='b',gridsize=25)
```

Compare this to a dataframe created using `.rand()`

```python
df = pd.DataFrame(np.random.rand(1000,2),columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=20)
```
