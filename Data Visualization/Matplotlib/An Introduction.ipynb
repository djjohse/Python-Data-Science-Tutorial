**Matplotlib has a functional and object oriented method of creating plots**

---

# Functional Method

`.plot()` takes two NumPy arrays and stores a plot in memory. There are additional functions that can be called for labeling. Finally, the resulting object is printed using `.show()`

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

`.subplot()` takes a 3-digit integer, where the first digit is the number of rows, the second the number of columns, and the third the index of the subplot. This 3-digit integer can be separated by commas.

```python
plt.subplot(121)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

---

# Object Oriented Method

`.figure()` requires manual creation of axes through the use of the method `.add_axes()`

`.add_axes()` takes a list of left, bottom, width, and height values. All quantities are in fractions of figure width and height.

Once a figure has axes, values can be ploted using `.plot()` as in the functional method.

Labels can be added by using
`.set_xlabel()`
`.set_ylabel()`
`.set_title()`

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')

plt.show()
```

## Multi-Plot

The first two arguments in `.add_axes()` relate to position.
The last two arguments relate to scale.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.2,0.7,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')

plt.show()
```

## `.subplot()`

```python
fig,axes = plt.subplots()

axes.plot(x,y)
plt.show()
```

`.subplots()` takes nrows and ncols as arguments to create a multi-plot

`.tight_layout()` can alternatively be used to `.show()` with the added benefit of fixing any overlapping.

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

Note: axes is just an Array of axis stored in memory. Therefore, it can be iterated through if necessary.

For example,

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

Or perhaps more importantly, the axes can be indexed

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.show()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

This works as well with `.subplots()`

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

# Legends

To add a legend, first your plot objects must be labeled. This is done by passing the label attribute to `.plot()`

Then, `.legend()` will read your labeled objects and add them to a legend.

For legend location options, read the documentation regarding the [loc](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html "matplotlib.pyplot.legend Documentation") common location codes, or pass in the coordinates to specify the location of the legend's bottom corner.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x-squared')
ax.plot(x,x**3, label='x-cubed')

ax.legend(loc=(0.2,0.6))

plt.show()
```

# Plot Appearance

The **color** attribute can accept common color strings, such as 'red' and 'purple', or can accept a RGB hex code.

The **linewidth**, or **lw**, attribute can adjust the thickness of the plot line.

The **alpha** attribute can adjust the transparency of the plot line.

**linestyle**, or **ls**, can adjust the way the plot line is rendered.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', lw=5, alpha=0.5, ls='-.')

plt.show()
```

Markers can be used to locate the position of the data points within the original arrays.
They have their own styling as well.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='o', markersize=12, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='red')

plt.show()
```

### Control over axis appearance

To display only a section of the total plot,

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,lw=2,marker='o')

ax.set_xlim([0,1])
ax.set_ylim([0,2])

plt.show()
```
