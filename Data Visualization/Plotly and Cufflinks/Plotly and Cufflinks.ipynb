### Data

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
df2
```

## Pandas `.plot()`

```python
df.plot()
```

## Plotly `.iplot()`

```python
df.iplot()
```

---

## `kind = scatter`

By default, Plotly tries to draw a line between points for the `.iplot` method. To bypass this feature, you can pass the argument **`mode='markers'`**

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

---

## `kind = bar`

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

Sometimes data requires some manipulation to create a bar plot.

For example, look at the bar plot for **df**

```python
df.iplot(kind='bar')
```

This gives bars for every single data point and is difficult to read.

To clean up the output for this, perform some aggregate function on top of `.iplot()`

```python
df.sum().iplot(kind='bar')
```

---

## `type = 'box'`

```python
df.iplot(kind = 'box')
```

## 3D Surface Plots

First let's create a new DataFrame

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

---

## `kind = 'hist'`

Plotly can overlap histograms so that they can be compared without having to select any particular columns.

```python
df.iplot(kind='hist',bins=30)
```

---

## `kind = 'spread'`

Graphs a line plot and a subplot that shows the difference, or 'spread', between the lines 

```python
df[['A','B']].iplot(kind='spread')
```

---

## `kind = 'bubble'`

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```
