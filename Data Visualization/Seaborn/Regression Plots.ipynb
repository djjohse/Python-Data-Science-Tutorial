```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

---

## `.lmplot()`

Refer to the [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html) for the linear model plot usage.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

Keywords can be passed to `plt.scatter` and `plt.plot` by use of the **`{scatter,line}_kws`** argument.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

Instead of passing a **`hue`** parameter, a second plot can be graphed by use of **`col`**.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

Seaborn figures can have size and aspect ratio adjusted with **`size`** and **`aspect`** parameters.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)
```
