```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

---

## `.set_style()`

You can set the background of your figures by calling `.set_style()`

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

---

## `.despine()`

You can remove the framing of the figure by calling `.despine()`

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

---

## `plt.figure()`

Certain features within matplotlib can be used to modify seaborn functions. In this example, `.figure()` from matplotlib is used to change the size of the figure.

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

---

## `.set_context()`

The context can be set to paper, notebook, talk, or poster to best suit the print media. Notebook is default.

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

---

## Color Maps

You can explore [colormaps_reference.py](https://matplotlib.org/examples/color/colormaps_reference.html) online for a list of available color maps.

```python
sns.set_context('notebook')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Dark2')
```
