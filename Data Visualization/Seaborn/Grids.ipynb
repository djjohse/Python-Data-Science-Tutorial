```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

Now that we've seen the head of this dataset, we can begin exploring it.

We can call `.unique()` on the species column to see what species exist in this dataset.

We can also attempt to get an overview by using `.pairplot()`

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

Using `.PairGrid` we can customize these plots with a lot more control.

```python
g = sns.PairGrid(iris)

g.map(plt.scatter)
```

With `.PairGrid` you have even more control over these plots.

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## `.FacetGrid()`

Multi-plot grid for plotting conditional relationships.

```python
tips = sns.load_dataset('tips')
tips.head()
```

Generally, you have three arguments in a Facet Grid. **`row`** and **`col`** are the variables that separate each graph to give you a unique plot. Then in your mapping function, you will have the variables you are interested in mapping for each graph.

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```
